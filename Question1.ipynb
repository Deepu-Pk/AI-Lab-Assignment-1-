```
import numpy as np
import matplotlib.pyplot as plt
W1 = 1
W2 = 1
W3 = 2
Xmin, Xmax = -10,10
x1 = np.linspace(Xmin,Xmax,100)
x2 = -(W3/W2)-((W1*x1)/W2)
plt.plot(x1,x2,color='blue')
XaxisMin,XaxisMax = -10,10
YaxisMin,YaxisMax = -10,10
plt.axis([XaxisMin,XaxisMax,YaxisMin,YaxisMax])
plt.fill_between(x1,x2,YaxisMax,color = 'green')
```
